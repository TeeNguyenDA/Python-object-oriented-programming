### **Question**:
Define three classes, `iLine`, `iCurved`, `iStraight`, where `iLine` is the parent of `iCurved` and `iStraight`.

**a)** The `iLine` should have the following methods:<br>
  * A method called `whoAmI` prints out `Curved Line` if the object has three coefficients: $a, b$ and $c$ or `Straight Line` if the object has only two coefficients: $m$ and $c$.<br>
  * A method called `lineEquation` prints the equation of the line:<br>
    * The `iCurved` equation is: $y = ax^2+bx+c$
    * The `iStraight` equation is $y = mx+c$

**b)**  When creating the `iCurved` objects, the program should generate three random integers as the coefficients of the line. Similarly, the program should randomly generate two random integers denoting the coefficients of the `iStraight` objects. The coefficients can be any value between -10 and 10, inclusively.

**c)** Each `iCurved` and `iStraight` class should have a method called `plotLine` to plot the respective shapes using the [Matplotlib Pyplot](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.html) function, and only the `iCurved` class has another method called `findRoots` that computes and print out the roots of the curve using the [Numpy Roots](https://numpy.org/doc/stable/reference/generated/numpy.roots.html) function.

The program should automatically create 3 Straight and 3 Curved objects and print out their properties.

```python
from random import randrange
import numpy as np
import matplotlib.pyplot as plt

class iLine:
    def __init__(self, no_of_coef):
        self.no_of_coef = no_of_coef
        self.coef = [0 for i in range(no_of_coef)]
        
    def whoAmI(self):
        if self.no_of_coef == 3:
            a, b, c = self.coef
            print("Curved Line")
            print("Number of Coefficients:", 3)            
        elif self.no_of_coef == 2:
            m, c = self.coef
            print("Straight Line")
            print("Number of Coefficients:", 2)
        else:
            print("Not what we want")
            
    def lineEquation(self, a, b, c, m=0):     
        if self.no_of_coef == 3:
            self.coef = (a, b, c)
            print("The equation of the Curved Line is: y = {a}(x^2) + {b}*x + {c}".format(a=a, b=b, c=c))
        elif self.no_of_coef == 2:
            self.coef = (m, c)
            print("The equation of the Straight Line is: y = {m}*x + {c}".format(m=m,  c=c))
        else:
            print("No equation to display")
   
    def plotLine(self, a, b, c, m=0): 
        if self.no_of_coef == 3:
            self.coef = (a, b, c)
            x = np.linspace(-2, 2, 100)
            y = np.poly1d([a, b, c])(x)
            plt.plot(x, y)
            plt.show()
        elif self.no_of_coef == 2:
            self.coef = (m, c)
            x = np.linspace(-2, 2, 100)
            y = np.poly1d([m, c])(x)
            plt.plot(x, y)
            plt.show()
        else:
            print("No plot to show")      

    def findRoots(self, a, b, c, m=0): 
        if self.no_of_coef == 3:
            self.coef = (a, b, c)
            print("The roots are:", "x1 =", np.roots(self.coef)[0], "and x2 =", np.roots(self.coef)[1])
        elif self.no_of_coef == 2:
            self.coef = (m, c)
            pass
        else:
            pass
    

class iCurved(iLine):
    def __init__(self):
        iLine.__init__(self, 3)
        
class iStraight(iLine):
    def __init__(self):
        iLine.__init__(self, 2)

# 2b:

# iCurved and iStraight - First Object:
first_a = int(randrange(-10, 10))
first_b = int(randrange(-10, 10))
first_c = int(randrange(-10, 10))
first_m = int(randrange(-10, 10))

iCurved().whoAmI()
iCurved().lineEquation(first_a, first_b, first_c, 0)  # lineEquation(a, b, c, m=0)
iCurved().findRoots(first_a, first_b, first_c, 0)     # findRoots(a, b, c, m=0)
iCurved().plotLine(first_a, first_b, first_c, 0)      # plotLine(a, b, c, m=0)


iStraight().whoAmI()
iStraight().lineEquation(0, 0, first_c, first_m)      # lineEquation(a, b, c, m=0)
iStraight().plotLine(0, 0, first_c, first_m)          # plotLine(a, b, c, m=0)

# iCurved and iStraight - Second Object:
second_a = int(randrange(-10, 10))
second_b = int(randrange(-10, 10))
second_c = int(randrange(-10, 10))
second_m = int(randrange(-10, 10))

iCurved().whoAmI()
iCurved().lineEquation(second_a, second_b, second_c, 0)  # lineEquation(a, b, c, m=0)
iCurved().findRoots(second_a, second_b, second_c, 0)     # findRoots(a, b, c, m=0)
iCurved().plotLine(second_a, second_b, second_c, 0)      # plotLine(a, b, c, m=0)

iStraight().whoAmI()
iStraight().lineEquation(0, 0, second_c, second_m)      # lineEquation(a, b, c, m=0)
iStraight().plotLine(0, 0, second_c, second_m)          # plotLine(a, b, c, m=0)

# iCurved and iStraight - Third Object:
third_a = int(randrange(-10, 10))
third_b = int(randrange(-10, 10))
third_c = int(randrange(-10, 10))
third_m = int(randrange(-10, 10))

iCurved().whoAmI()
iCurved().lineEquation(third_a, third_b, third_c, 0)  # lineEquation(a, b, c, m=0)
iCurved().findRoots(third_a, third_b, third_c, 0)     # findRoots(a, b, c, m=0)
iCurved().plotLine(third_a, third_b, third_c, 0)      # plotLine(a, b, c, m=0)

iStraight().whoAmI()
iStraight().lineEquation(0, 0, third_c, third_m)      # lineEquation(a, b, c, m=0)
iStraight().plotLine(0, 0, third_c, third_m)          # plotLine(a, b, c, m=0)
```
